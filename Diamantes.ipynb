Análise da Estatística Descritiva / Gráfica de todas as variáveis Quanlitativas (Ordinal) do table



```python
print("Análise Descritiva das Variáveis Qualitativas Ordinais")

pd.set_option("display.max_rows", None)
pd.set_option("display.max_columns", None)

# Ordem lógica para "CUT"
cut_order = ["High Quality", "Low Quality"]

# Ordem padrão GIA de 'melhor' para 'pior' para "COLOR"
color_order = ['D', 'E', 'F', 'G', 'H', 'I', 'J']

# Ordem padrão GIA de 'melhor' para 'pior' para "CLARITY"
clarity_order = ['IF', 'VVS1', 'VVS2', 'VS1', 'VS2', 'SI1', 'SI2', 'I1']

# Dicionário para armazenar as variáveis e suas ordens
ordinal_vars = {
        'CUT': cut_order,
        'COLOR': color_order,
        'CLARITY': clarity_order
    }

print("Análise Descritiva das Variáveis Qualitativas Nominais (Tabelas de Frequência)")

# Geração das análises descritivas (frequência e porcentagem) para cada
for col in ordinal_vars.keys():
  print(f"Análise Descritiva para a coluna: {col}")

  # Calcular contagem
  counts = table[col].value_counts()

  # Calcular porcentagem
  percentages = table[col].value_counts(normalize=True).mul(100).round(2)

  # Criação da tabela de frequência
  freq_table = pd.DataFrame({
      "Contagem (Frequência)" : counts,
      "Porcentagem (%)": percentages
  })

  # Ordenação pela ordem ordinal definida com utilização de .reindex
  freq_table = freq_table.reindex(ordinal_vars[col])

  print(freq_table)

# Análise Gráfica
print("Análise Gráfica das Variáveis Qualitativas Ordinais")

for col, order in ordinal_vars.items():
  print(f"Gerando gráfico para: {col}")

  # Garantia de que todas as categorias estarão na ordem que foram definidas
  actual_order = [cat for cat in order if cat in table[col].unique()]

  # Criação do gráfico de barras
  sns.countplot(data=table, x=col, order=actual_order)

  plt.title(f"Distribuição Ordinal da Variável: {col}", fontsize = 16)
  plt.ylabel("Contagem(Frequência)", fontsize=12)
  plt.xlabel(f"Categoria de {col} (em ordem)", fontsize=12)

  # Rotação do texto do eixo X
  if col == "CLARITY" or col == "COLOR":
    plt.xticks(rotation=45)

  plt.tight_layout()
  plt.show()

# Análises separada por Grupo de "CUT"
df_low = table[table['CUT'] == 'Low Quality'].copy()
df_high = table[table['CUT'] == 'High Quality'].copy()

# Lista dos grupos para análise
datasets_para_analisar = [
    ("Low Quality", df_low),
    ("High Quality", df_high)
]

# Dicionário de variáveis dentro dos grupos
vars_subgrupo = {
    'COLOR': color_order,
    'CLARITY': clarity_order
}

# Tabelas e Gráficos (por grupo)

for group_name, group_df in datasets_para_analisar:

    print(f"\nIniciando análise do grupo: {group_name.upper()}")

    # Tabelas de Frequência (por grupo)
    print(f"\nTabelas de Frequência ({group_name})")

    # Loop interno
    for col in vars_subgrupo.keys():
        print(f"\nAnálise Descritiva para: {col} (Grupo: {group_name})")

        counts = group_df[col].value_counts()
        percentages = group_df[col].value_counts(normalize=True).mul(100).round(2)
        freq_table = pd.DataFrame({
            "Contagem (Frequência)" : counts,
            "Porcentagem (%)": percentages
        })
        freq_table = freq_table.reindex(vars_subgrupo[col])
        print(freq_table)

    print(f"\nGráficos ({group_name})")

    # Loop interno
    for col, order in vars_subgrupo.items():
        print(f"Gerando gráfico para: {col} (Grupo: {group_name})")

        actual_order = [cat for cat in order if cat in group_df[col].unique()]

        plt.figure()
        sns.countplot(data=group_df, x=col, order=actual_order)

        # Título dinâmico incluindo o nome do grupo
        plt.title(f"Distribuição de {col} (Grupo: {group_name})", fontsize=16)
        plt.ylabel("Contagem(Frequência)", fontsize=12)
        plt.xlabel(f"Categoria de {col} (em ordem)", fontsize=12)

        if col == "CLARITY" or col == "COLOR":
            plt.xticks(rotation=45)

        plt.tight_layout()
        plt.show()
```

#  Análise de Outliers por Grupo de Qualidade





```python
# Definição dos grupos para análise
groups = ["Low Quality", "High Quality"]

# Lista para armazenar os resultados da contagem de outliers
outlier_results = []

# Cálculo e Contagem de Outliers
print("Calculando Outliers pelo Método IQR por Grupo")

# Loop por grupo
for group in groups:
    group_df = table[table['CUT'] == group]
    total_rows_group = len(group_df)

    # Loop por cada colunas quantitativa
    for col in continuous_cols:

      # Cálculo Q1 e Q3
      Q1 = group_df[col].quantile(0.25)
      Q3 = group_df[col].quantile(0.75)

      # Cálculo IQR
      IQR = Q3 - Q1

      # Definição dos limites (fences)
      lower_bound = Q1 - (1.5 * IQR)
      upper_bound = Q3 + (1.5 * IQR)

      # Identificação e contagem dos outliers
      outlier_mask = (group_df[col] < lower_bound) | (group_df[col] > upper_bound)
      outlier_count = group_df[col][outlier_mask].count()

      # Armazenamento dos resultados
      outlier_results.append({
          "Grupo": group,
          "Coluna": col,
          "Total de Linhas": total_rows_group,
          "Total de Outliers": outlier_count,
          "Porcentagem de Outliers": (outlier_count / total_rows_group) * 100,
          "Limite Inferior": f"{lower_bound:.2f}",
          "Limite Superior": f"{upper_bound:.2f}"
      })

# Criação de um DataFrame com os resultados e impressão
df_results = pd.DataFrame(outlier_results)

print("Tabela de Contagem de Outliers pelo Método IQR")
print(df_results.to_string())

# Análise Gráfica Comparativa por Grupo
for col in continuous_cols:
  sns.boxplot(data=table, x="CUT", y=col, order=["Low Quality", "High Quality"])
  plt.title(f"Distribuição de {col} por Grupo", fontsize=12)
  plt.xlabel("Grupo(CUT)", fontsize=12)
  plt.ylabel(f"Valor de {col}", fontsize=12)
  plt.tight_layout()
  plt.show()
```

# Conclusões

### Insights sobre o grupo "CUT: Low Quality"

*   Diamantes "Low Quality" com cores D, E ou F (quase 75.000 unidades) podem ser comercializados com uma excelente relação custo-benefício.
*   Os diameantes de 2.50 quilates com clareza SI1/VS2 são os mais frequentes. As campanhas devem focar na disponibilidade e no preço competitivo deste exato perfil.
*   O grupo "Low Quality" é o produto de maior volume. Qualquer estratégia de redução de inventário ou promoção deve começar por esses, pois é onde o maior capital está alocado (75% das unidades totais).
*   Definição de faixa de preço 'normal' para este grupo entre 4.032,00 a 7.080,00 dólares, sendo que apenas 1,34% desse grupo está fora dessa faixa.
*   Ao analisar as dimensões desses diamantes, é possível afirmar que a maior fonte de outliers neste grupo não é o preço ou o tamanho, mas suas proporções. Há uma chance maior de um diamante deste grupo ser fisicamente desproporcional(TABLE ou DEPTH) do que ser um outlier de tamanho (CARAT).
*   A análise sugere uma forte correlaçao de que os preços mais baixos são os outliers desproporcionais, sendo esses o maior risco de stock. A estratégia deveria ser de liquidar esses itens com descontos ou vendê-los em lotes.

### Insights sobre o grupo "CUT: High Quality"

*   O cliente "High Quality" paga 72% a mais pelo mesmo quilate, cor e clareza de um cliente "Low Quality". A diferença crucial é o brilho, ou seja, a performance visual da pedra.
*   Os dados apontam que o desvio padrão no grupo "High Quality" é baixo ($689), o que significa que o preço é bem aceito pelo mercado. É possível inferir que não há necessiadde de descontos neste grupo e que isso pode destruir a percepção de valor do produto.
*   O "High Quality" representa apenas 25% do inventário, sendo raro e exclusivo. É válido usar este como o protagonista nas campanhas de marketing e vitrines.
*   A faixa de preço é extremamente definida: entre 8.056,50 a 11.076,50 dólares. Apenas 1,43% dos diamantes deste grupo estão fora dessa faixa.
*   O risco físico também afeta este grupo. Por mais que o tamanho (CARAT) seja consistente, as proporções físicas são a maior fonte de outliers.
*   Os outliers superiores em faixa de preço são uma grande oportunidade e podem se tornar flagships. Podendo ser comercializados como a elite do inventário.
